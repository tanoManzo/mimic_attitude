```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import json
#add json files

BLOOM_PATH = "/content/drive/MyDrive/mitt/urop/plotting/log_bloom.json"
DISTILBERT_PATH = "/content/drive/MyDrive/mitt/urop/plotting/log_distilbert.json"
MINILM_PATH = "/content/drive/MyDrive/mitt/urop/plotting/log_miniLM.json"
ROBERTA_PATH = "/content/drive/MyDrive/mitt/urop/plotting/log_roberta.json"

```

```python
df_bloom_expanded = pd.read_json(BLOOM_PATH)
df_distilbert_expanded = pd.read_json(DISTILBERT_PATH)
df_minilm_expanded = pd.read_json(MINILM_PATH)
df_roberta_expanded = pd.read_json(ROBERTA_PATH)

df_roberta_expanded
```

```python

df_bloom = df_bloom_expanded.groupby(df_bloom_expanded['epoch']).aggregate('max')
df_distilbert = df_distilbert_expanded.groupby(df_distilbert_expanded['epoch']).aggregate('max')
df_minilm = df_minilm_expanded.groupby(df_minilm_expanded['epoch']).aggregate('max')
df_roberta = df_roberta_expanded.groupby(df_roberta_expanded['epoch']).aggregate('max')

df_roberta
```

```python
#df_bloom['loss']
```

```python
def plot_model(df, attribute):
  fig, ax = plt.subplots()
  x = [x for x in range(1,101)]
  y = df[attribute]

  ax.set_xlabel("Epochs")
  ax.set_ylabel(attribute)
  
  plt.title(attribute)
  ax.plot(x, y)
  plt.show()
```

```python
plot_model(df_bloom, 'loss')
plot_model(df_bloom, 'eval_loss')
```

```python
plot_model(df_distilbert, 'loss')
plot_model(df_distilbert, 'eval_loss')
```

```python
plot_model(df_minilm, 'loss')
plot_model(df_minilm, 'eval_loss')
```

```python
plot_model(df_roberta, 'loss')
plot_model(df_roberta, 'eval_loss')
```

```python

#take in list of dfs
def plot_all(dfs, attribute):
  for frame in dfs:
    #not sure how/if can print name of each df in this format
    plot_model(frame, attribute)
```

```python
models = [df_bloom, df_distilbert, df_minilm, df_roberta]
#loss	eval_loss	eval_f1_weighted	eval_f1_macro	eval_f1_micro	eval_accuracy	eval_runtime
plot_all(models, 'eval_f1_weighted')
```
